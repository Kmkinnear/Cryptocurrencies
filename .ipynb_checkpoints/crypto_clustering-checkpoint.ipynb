### Deliverable 1: Preprocessing the Data for PCA

```python
# Load the crypto_data.csv dataset.
#http://net-informations.com/ds/err/unnamed.htm#:~:text=There%20are%20situations%20when%20an%20Unnamed%3A%200%20column,reads%20in%20the%20first%20column%20as%20the%20index.
file_path = "crypto_data.csv"
df = pd.read_csv(file_path, index_col=[0])

df.head()
```

```python
# Keep all the cryptocurrencies that are being traded.
is_trading_df = df[df['IsTrading']==True] 
print(is_trading_df.shape)
is_trading_df.head(10)
```

```python
# Remove the "IsTrading" column. 
crypto_df = is_trading_df.drop(['IsTrading'], axis=1)
print(crypto_df.shape)
crypto_df.head(10)
```

```python
# Remove rows that have at least 1 null value.
# Drop null rows
crypto_df = crypto_df.dropna()
print(crypto_df.shape)
crypto_df.head(10)
```

```python
# Keep the rows where coins are mined.
crypto_df = crypto_df[crypto_df['TotalCoinsMined']>=1]
print(crypto_df.shape)
crypto_df.head(10)
```
